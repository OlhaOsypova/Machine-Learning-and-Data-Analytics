## Лабораторна робота 2  
У роботі використовується набір даних із щорічного опитування розробників, яке проводить сайт Stack Overflow. Це масштабне опитування охоплює десятки тисяч IT-спеціалістів з усього світу.
У датасеті збережені відповіді респондентів на різні питання: вік, освіта, кількість років програмування, рівень доходу, використання технологій, відношення до штучного інтелекту, досвід роботи, задоволеність роботою тощо.

### Завдання  
1. Огляд датасету  
- Завантажте дані в датафрейм.
- Виведіть загальну інформацію про датафрейм, зробіть короткі висновки. 
2. Видалення стовпців із великою кількістю пропусків  
- Проаналізуйте скільки пропусків є в даних.  
- Видаліть стовпці, де кількість пропусків перевищує 30%.
3. Аналіз та видалення викидів
- Використайте відомі вам методи для виявлення аномально великих або малих значень у числових змінних. Зробіть висновки щодо наявності викидів в даних.
- Видаліть викиди, якщо на попередньому кроці вони були виявлені в даних.
4. Обробка пропусків.  
- Перевірте, чи залишилися ще пропуски в даних.
- Якщо залишилися - заповніть пропуски за допомогою SimpleImputer.
5. Аналіз респондентів з України

а) Використовуючи змінні AISelect (чи використовує респондент AI) та AISent (ставлення до AI), проаналізуйте розподіл відповідей респондентів(<span style="color:red"> для цього необхідно самостійно опанувати метод value_counts()</span>).  
б) Зробіть короткий висновок про те, як українські IT-спеціалісти сприймають штучний інтелект (наприклад: який відсоток використовує AI, яке переважає ставлення — позитивне, нейтральне чи негативне).
- Згрупуйте респондентів за змінною MainBranch (основна сфера діяльності) та розрахуйте середній вік іт спеціалістів в кожній категорії. Чи можна помітити певні відмінності у віці? (<span style="color:red"> для цього необхідно самостійно опанувати метод groupby()</span>)

Зробіть висновки: чи існує різниця в досвіді між тими, хто працює віддалено, і тими, хто працює в офісі (<span style="color:red"> для цього необхідно самостійно опанувати метод groupby()</span>).
6. Кодування та стандартизація змінних (завдання виконуєте на повному наборі даних!!!)  
- Перекодуйте категоріальні змінні у числовий формат.
- Для числових змінних виконайте стандартизацію (наприклад, StandardScaler), щоб привести їх до єдиної шкали.
