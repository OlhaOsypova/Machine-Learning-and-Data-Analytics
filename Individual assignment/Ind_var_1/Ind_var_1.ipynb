## Досліджуємо вплив використання соціальних мереж на емоційний стан користувачів  
На основі даних про активність користувачів у соціальних мережах та їх домінантний емоційний стан дослідити можливість прогнозування емоційного стану користувачів залежно від характеру їхньої активності.  

**Опис датасету:**  
User_ID: Унікальний ідентифікатор користувача.  
Age: Вік користувача.  
Gender: Стать користувача (Female, Male, Non-binary).  
Platform: Соціальна мережа, якою користується користувач (Instagram, Twitter, Facebook, LinkedIn, Snapchat, Whatsapp, Telegram).  
Daily_Usage_Time (minutes): Щоденний час користування платформою в хвилинах.  
Posts_Per_Day: Кількість постів за день.  
Likes_Received_Per_Day: Кількість отриманих вподобайок за день.  
Comments_Received_Per_Day: Кількість отриманих коментарів за день.  
Messages_Sent_Per_Day: Кількість надісланих повідомлень за день.  
Dominant_Emotion: Домінантний емоційний стан користувача протягом дня (наприклад, Happiness, Sadness, Anger, Anxiety, Boredom, Neutral).  

**Завдання**
На основі аналізу та обробки даних у датасеті, визначте залежність між активністю користувачів у соцмережах та їхнім емоційним станом( <span style="color:red"> Особливу увагу зверніть на колонку Dominant_Emotion, яка містить декілька можливих значень емоційного стану користувача</span>):  
- Виконайте попередню обробку даних, враховуючи можливі проблеми з даними.  
- Оберіть алгоритм машинного навчання для прогнозування домінантного емоційного стану користувача, враховуючи особливості цільової змінної.
- Навчіть кілька алгоритмів для прогнозування домінантного емоційного стану користувача та порівняйте їх якість. Оберіть найкращий алгоритм для подальшої роботи.  
- Використайте обраний алгоритм для прогнозування емоційного стану для вибраних користувачів (інформація про них міститься в 'users_to_predict.csv'). 

**У висновках надайте:**  
- Опис використаних вами підходів до аналізу та обробки даних.  
- Обрану модель (алгоритм) і аргументацію вибору.  
- Показники якості реалізованих алгоритмів та результати прогнозування на тестовій вибірці.  
- Результати прогнозування емоційного стану для вибраних користувачів.
