## Хто лишається в грі: аналіз поведінки та прогноз залученості гравців  
Ігрова індустрія щодня генерує терабайти подій: сесії, години в грі, покупки, відгуки, переходи між жанрами. Метою вашої індивідуальної роботи є зрозуміти що утримує гравця в грі. Для цього ви фокусуєтеся на цільовій змінній EngagementLevel як індикаторі залученості. Ваша мета — виявити поведінкові та контекстні характеристики, що пов’язані з утриманням, і навчитися прогнозувати рівень залученості залежно від історії гри, жанру та активності, аби сформувати практичні дії для підвищення утримання гравця в грі. 
Набір даних містить інформацію про характеристики гравців, контекст гри та показники їхньої активності/залученості, що дозволяє дослідити чинники утримання та побудувати прогноз цільової змінної.
- PlayerID — унікальний ідентифікатор гравця.  
- Age — вік гравця (роки).  
- Gender — стать гравця.  
- Location — географічне розташування гравця.  
- GameGenre — жанр гри, у яку залучений гравець.  
- PlayTimeHours — середня кількість годин за одну сесію.  
- InGamePurchases — факт  покупок в грі (0 = ні, 1 = так).  
- GameDifficulty — рівень складності гри.  
- SessionsPerWeek — кількість ігрових сесій на тиждень.   
- AvgSessionDurationMinutes — середня тривалість сесії (хвилини).    
- PlayerLevel — поточний рівень гравця у грі.  
- AchievementsUnlocked — кількість здобутих досягнень.  
- EngagementLevel — цільова змінна: категоризований рівень залученості/утримання гравця (High, Medium, Low).

### Завдання  
1. Підготовка даних  
- Перевірте типи змінних, пропуски, аномальні значення; зафіксуйте кількість рядків/стовпців. За потреби заповніть пропуски, видаліть аномальні значення.  
- Видаліть службові змінні.

2. Знайомство з даними  
- Опишіть розподіл EngagementLevel (частки High/Medium/Low) і базові статистики для ключових числових ознак (PlayTimeHours, SessionsPerWeek, AvgSessionDurationMinutes, PlayerLevel, AchievementsUnlocked).
- Покажіть топ-жанри за кількістю гравців і як у них розподіляється EngagementLevel. Зробіть короткі висновки.  

3. Порівняння двох жанрів  
Оберіть два різні значення GameGenre і порівняйте профілі їхніх гравців за наявними характеристиками та за розподілом EngagementLevel. Зробіть короткий висновок: чим відрізняються аудиторії жанрів з огляду на залученість.  

4. Групування гравців   
Виконайте групування гравців за схожістю поведінкових показників, використавши числові ознаки (наприклад, PlayTimeHours, SessionsPerWeek, AvgSessionDurationMinutes, PlayerLevel, AchievementsUnlocked, за потреби — числове кодування GameDifficulty) і виключивши EngagementLevel. Охарактеризуйте отримані групи: розмір, типові значення ключових ознак і дайте інтерпретовані назви.  За можливості уникайте домінування бінарних індикаторів (як-от InGamePurchases).  

5. Модель прогнозування цільової змінної  
Побудуйте модель для прогнозування змінної EngagementLevel. Перевірте якість побудованої моделі та зробіть висновок щодо придатності моделі для подальшого використання.  

6. Аналіз факторів впливу  
Виконайте аналіз внеску ознак у прогнозування змінної EngagementLevel. Виділіть 3–5 характеристик, що найбільше впливають на належність до High/Medium/Low, і коротко інтерпретуйте: як зміни в поведінці гравців (години, частота сесій, рівень, досягнення, складність) пов’язані з підвищенням/зниженням залученості.
