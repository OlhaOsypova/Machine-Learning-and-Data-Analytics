- Знайдіть середню кількість годин, витрачених на навчання, для кожного рівня навчання.

Цей датасет представляє собою інформацію, зібрану в рамках опитування студентів технічних спеціальностей. Нижче наведено опис кожного стовпця датасету:  
**1. Student_identifier_code_for_participation_in_the_survey** - Ідентифікаційний код студента, який брав участь в опитуванні. Це унікальний код для кожного учасника.  
**2. Age_of_student_at_the_time_of_the_data_collection_process** - Вік студента на момент збору даних. Це кількісне значення, що варіюється в межах від 18 до 25 років.  
**3. Grade_level_of_the_participant_in_the_educational_program** - Рівень навчання студента в освітній програмі. Це категоріальна змінна, яке може приймати одне з наступних значень (категорій):  
- Freshman (перший курс),  
- Sophomore (другий курс),  
- Junior (третій курс),  
- Senior (четвертий курс).   

**4. Hours_spent_studying_on_a_weekly_basis** - Кількість годин, витрачених студентом на навчання протягом тижня. Це кількісне значення, яке може варіюватися від 5 до 40 годин.   

**5. Preferred_study_method_in_lecture_or_practical_sessions** - Обраний студентом метод навчання. Можливі варіанти:  
- Lecture (лекції),  
- Practical (практика),  
- Self-study (самостійне навчання),  
- Group-study (групове навчання).    

**6. Number_of_absent_days_in_the_current_semester** - Кількість пропущених днів у поточному семестрі. Це кількісне значення, яке може варіюватися від 0 до 15.  
**7. Average_test_score_in_mathematical_subjects** - Середній бал студента з математичних предметів. Це кількісне значення у діапазоні від 50 до 100 балів.  
**8. Student_participation_in_extracurricular_activities** - Участь студента в позанавчальних активностях. Це категоріальне значення, яке може приймати значення "Yes" (так) або "No" (ні).  
**9. Preferred_programming_language_for_project_work** - Обрана студентом мова програмування для виконання проектних робіт. Можливі варіанти:  
- Python,  
- C++,  
- Java,  
- JavaScript,  
- R.    

**10. Satisfaction_level_with_academic_facilities_in_percentage** - Рівень задоволеності навчальним закладом, виражений у відсотках. Це кількісне значення від 50% до 100%.    

**11. Plans_for_further_education_or_employment_after_graduation** - Плани студента після завершення навчання. Це категоріальна змінна, яке може приймати такі зачення:  
- Higher education (продовження навчання),  
- Employment (працевлаштування),  
- Freelancing (фріланс),  
- Undecided (не визначився).
