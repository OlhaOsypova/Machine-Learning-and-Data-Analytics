## Лабораторна робота 1: обробка даних із використанням бібліотеки Pandas  
Початкові дані містять інформацію про 1200 студентів, зібрану під час опитування.  
У таблиці є наступні стовпчики:  
- Student_identifier_code_for_participation_in_the_survey – унікальний ідентифікатор студента.  
- Age_of_student_at_the_time_of_the_data_collection_process – вік студента під час збору даних.  
- Grade_level_of_the_participant_in_the_educational_program – рівень навчання (Freshman, Sophomore, Junior, Senior).  
- Hours_spent_studying_on_a_weekly_basis – кількість годин, які студент витрачає на навчання протягом тижня.  
- Preferred_study_method_in_lecture_or_practical_sessions – улюблений метод навчання (Lecture, Practical, Self-study).  
- Number_of_absent_days_in_the_current_semester – кількість пропущених занять у поточному семестрі.  
- Average_test_score_in_mathematical_subjects – середній бал з математичних дисциплін.  
- Student_participation_in_extracurricular_activities – участь у позанавчальній діяльності (Yes/No).  
- Preferred_programming_language_for_project_work – мова програмування, яку студент віддає перевагу для проєктів.  
- Satisfaction_level_with_academic_facilities_in_percentage – рівень задоволеності навчальною інфраструктурою (у відсотках).  
- Plans_for_further_education_or_employment_after_graduation – плани після випуску (Employment, Higher Education, Undecided).  

### Завдання до ЛР  
Використовуючи можливості бібліотеки Pandas виконайте наступні кроки:  
1. Завантаження даних
- Зчитайте вміст файлу із даними у датафрейм.
- Виведіть перші 5 рядків таблиці, щоб побачити, як виглядають записи.
2. Огляд структури даних  
- Визначте загальну кількість рядків і стовпців.
- Подивіться типи даних у кожному стовпці.
3. Обробка пропусків
- Знайдіть, у яких стовпцях є пропуски, і скільки пропущених значень у кожному.
- Видаліть усі рядки, які мають хоча б один пропуск.
4. Перейменування стовпчиків
- Зробіть імена стовпчиків більш зручними та короткими. Прикладами можуть бути: Student_ID замість Student_identifier_code_for_participation_in_the_survey і т.д. 
5. Основна описова статистика числових змінних
- Використайте метод describe() для кількісних стовпчиків.
- Коротко (у вигляді кількох речень) опишіть, що можна сказати про змінні вік студентів та бал з математичних дисциплін.  
6. Основна описова статистика категоріальних змінних
- Використайте метод describe(include=['object']) для категоріальних стовпців.
- Проаналізуйте результати: які рівні навчання найпоширеніші, яка мова програмування обирається найчастіше тощо.
- Напишіть короткий опис (2–3 речення) ваших спостережень.
7. Фільтрація та отримання зрізів
- Виберіть усіх студентів, які навчаються більше ніж 20 годин на тиждень. Виведіть лише ті рядки, які відповідають цій умові, без використання loc. 
Поясніть, як працює фільтрація через логічний вираз.
- Виберіть дані лише тих студентів, які мають середній бал з математики вище 75 та для цих студентів виведіть тільки стовпці з унікальним ідентифікатором студента, віком та балом з математики.  Поясніть, чому тут зручніше використовувати loc, а не просту фільтрацію.
- Виберіть студентів, які одночасно: витрачають на навчання більше ніж 20 годин на тиждень та їх задоволеність навчальною інфраструктурою більше ніж 70%. Для цих студентів виведіть лише стовпці, що містять інформацію про к-ть годин навчання, улюблений метод навчання та рівень задоволеності навчальною інфраструктурою.
8. Збережіть датафрейм з перейменованими стовпчиками та видаленими пропусками в csv файл.
